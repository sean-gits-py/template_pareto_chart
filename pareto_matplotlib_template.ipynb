```python
import pandas as pd
#import plotly as plt
```

```python
### STEP 1 -> Generate Data ###

#create dataframe
df = pd.DataFrame({'count': [506, 120, 86, 34, 20, 18, 15, 11, 10, 9]})
df.index = [ 'A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
```

```python
#sort dataframe by count descending
df = df.sort_values(by='count', ascending=False)
```

```python
#create cumulative percentage column

df['cumulative_percent'] = df['count'].cumsum()/df['count'].sum()*100
```

```python
df
```

```python
### STEP 2 -> Generate MATPLOTLIB Chart ###
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

#specify plot color and line size
color1 = 'steelblue'
color2 = 'gold'
line_size = 4
```

```python
#create basic bar plot
fig, ax = plt.subplots()
ax.bar(df.index, df['count'], color=color1)
```

```python
#add cumulative percentage line to plot
ax2 = ax.twinx()
ax2.plot(df.index, df['cumulative_percent'], color=color2, marker="D", ms=line_size)
ax2.yaxis.set_major_formatter(PercentFormatter())

#specify axis colors
ax.tick_params(axis='y', colors=color1)
ax2.tick_params(axis='y', colors=color2)
```
